```python
import numpy as np #linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt # Importing the matplotlib library for plotting
import seaborn as sns # Importing the seaborn library for statistical data visualization
```

```python
#import the dataset
df = pd.read_csv("wine_data.csv", usecols= [0,1,2])
df.head()
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
sns.kdeplot(df['alcohol'], ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
sns.kdeplot(df['malic_acid'],ax=ax)
```

```python

# Define the color dictionary
color_dict = {1: 'red', 2: 'blue', 3: 'green'}

# Create the scatter plot
sns.scatterplot(data=df, x='alcohol', y='malic_acid', hue='class_label', palette=color_dict)

# Show the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the DataFrame into features (X) and target variable (y)
X = df.drop('class_label', axis=1)  # Features: drop the 'class_label' column
y = df['class_label']                 # Target variable: 'class_label'

# Use train_test_split to split the data
X_train, X_test, y_train, y_test = train_test_split(
    X,          # Features
    y,          # Target variable
    test_size=0.3,  # Proportion of the dataset to include in the test split
    random_state=0  # Seed for reproducibility
)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)

# Transform the test data (do not fit again)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
np.round(X_train.describe())
```

```python
# Create a figure with two subplots arranged in one row (ncols=2)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))  
# figsize sets the width and height of the figure in inches

# Create a scatter plot on the first subplot (ax1) using training data before scaling
ax1.scatter(X_train['alcohol'], X_train['malic_acid'], c=y_train)  
# X_train['alcohol'] is plotted on the x-axis, X_train['malic_acid'] on the y-axis
# c=y_train colors the points based on their class labels

# Set the title for the first subplot
ax1.set_title("Before Scaling")  

# Create a scatter plot on the second subplot (ax2) using scaled training data
ax2.scatter(X_train_scaled['alcohol'], X_train_scaled['malic_acid'], c=y_train)  
# X_train_scaled['alcohol'] is plotted on the x-axis, X_train_scaled['malic_acid'] on the y-axis
# c=y_train colors the points based on their class labels

# Set the title for the second subplot
ax2.set_title("After Scaling")  

# Display the plots
plt.show()  
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))  
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['alcohol'],ax = ax1 )
sns.kdeplot(X_train['malic_acid'],ax = ax1)


ax1.set_title("After Scaling")
sns.kdeplot(X_train_scaled['alcohol'],ax = ax2 )
sns.kdeplot(X_train_scaled['malic_acid'],ax = ax2)
plt.show()
```

```python
# Create a figure with two subplots arranged in one row (ncols=2)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))  
# figsize sets the width and height of the figure in inches

# Set the title for the first subplot (before scaling)
ax1.set_title("Before Scaling")  

# Create a Kernel Density Estimate (KDE) plot for 'alcohol' on the first subplot (ax1)
sns.kdeplot(X_train['alcohol'], ax=ax1)  
# This plots the density of the 'alcohol' feature in the training set

# Create a KDE plot for 'malic_acid' on the same subplot (ax1)
sns.kdeplot(X_train['malic_acid'], ax=ax1)  
# This adds the density of the 'malic_acid' feature to the same plot

# Set the title for the second subplot (after scaling)
ax2.set_title("After Scaling")  

# Create a KDE plot for scaled 'alcohol' on the second subplot (ax2)
sns.kdeplot(X_train_scaled['alcohol'], ax=ax2)  
# This plots the density of the scaled 'alcohol' feature in the training set

# Create a KDE plot for scaled 'malic_acid' on the same subplot (ax2)
sns.kdeplot(X_train_scaled['malic_acid'], ax=ax2)  
# This adds the density of the scaled 'malic_acid' feature to the same plot

# Display all created plots
plt.show()  
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5)) 
ax1.set_title("Alcohole distribution Before Scaling")
sns.kdeplot(X_train['alcohol'], ax=ax1)


ax2.set_title("Alcohole distribution after standard Scaling")
sns.kdeplot(X_train_scaled['alcohol'], ax=ax2)

```
