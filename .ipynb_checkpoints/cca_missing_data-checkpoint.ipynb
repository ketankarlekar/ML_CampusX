```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_science_job.csv")
```

```python
df.sample(5)
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
len(df[cols].dropna()) / len(df) 
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
new_df.hist(bins=50,density = True, figsize = (12,12))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

#original data
df['training_hours'].hist(bins=50,density = True, figsize = (12,12), ax = ax, color = "red")

new_df['training_hours'].hist(bins=50,density = True, figsize = (12,12), ax = ax, color = "green",alpha = 0.8)
```

```python


# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 12))

# Plot histograms
df['training_hours'].hist(bins=50, density=True, ax=ax, color="red", alpha=0.5, label='Original Data')
new_df['training_hours'].hist(bins=50, density=True, ax=ax, color="green", alpha=0.5, label='New Data')

# Add title and labels
ax.set_title('Distribution of Training Hours')
ax.set_xlabel('Training Hours')
ax.set_ylabel('Density')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
fig =plt.figure()
ax = fig.add_subplot(111)
#orignal data
df['training_hours'].plot.density(color = 'red')
#data after cca
new_df['training_hours'].plot.density(color = 'green')
```

```python


# Calculate percentage of observations per category for each dataset
original_dist = df["enrolled_university"].value_counts(normalize=True)
new_dist = new_df["enrolled_university"].value_counts(normalize=True)

# Concatenate the distributions into a DataFrame
temp = pd.concat([original_dist, new_dist], axis=1)

# Add column names
temp.columns = ["original", "new_data"]

# Display the resulting DataFrame
print(temp)
```
